## **2. Transformación y Clasificación Jerárquica (Data Transformation)**
Se aplican protocolos de ingeniería de datos para estandarizar la información. Se implementa una doble granularidad (Categoría y Herramienta) para permitir análisis estadísticos precisos:
* **Traducción Automática (NLP):** Localización de nombres de cursos y descripción de carga horaria al español.
* **Calidad de Datos:** Normalización de la variable temporal manteniendo el contexto original traducido.
* **Análisis Jerárquico:** Clasificación automatizada por Dominio de Aplicación y Especialidad Técnica, priorizando la detección de Inteligencia Artificial.

```python
import re

def corregir_terminologia(texto):
    """Protege términos técnicos contra traducciones erróneas post-traductor."""
    if not isinstance(texto, str): return texto
    diccionario_soporte = {
        r'\bpitón\b': 'Python',
        r'\babrazando la cara\b': 'Hugging Face',
        r'\bvaya\b': 'Go',
        r'\bir\b': 'Go',
        r'\bflujo de estado\b': 'Stateflow',
        r'\brampa de entrada\b': 'Onramp',
        r'\brampa de acceso\b': 'Onramp'
    }
    for patron, reemplazo in diccionario_soporte.items():
        texto = re.sub(patron, reemplazo, texto, flags=re.IGNORECASE)
    return texto

def clasificar_datos_pro(nombre):
    nombre_low = str(nombre).lower()

    # 1. INTELIGENCIA ARTIFICIAL (Prioridad Máxima)
    es_ia = any(kw in nombre_low for kw in [
        'intelligence', 'learning', 'ml', 'generative', 'prompt',
        'transformers', 'bert', 'neural', 'hcia-ai', 'agentes',
        'aprendizaje', 'generativo', 'redes neuronales',
        'ética de la ia', 'elementos de la ia', 'búsqueda e ia', 'general de la ia',
        'atención', 'incrustaciones', 'procesamiento del lenguaje', 'visión por computadora'
    ]) or re.search(r'\b(ia|ai)\b', nombre_low)

    if es_ia:
        return pd.Series(["Inteligencia Artificial y Machine Learning", 'Inteligencia Artificial'])

    # 2. HERRAMIENTAS TÉCNICAS (Prioridad 2: Captura el "¿Con qué?")
    if 'python' in nombre_low:
        return pd.Series(["Ciencia y Análisis de Datos (Data Science, Analytics, SQL)", 'Python'])
    if re.search(r'\b(r)\b', nombre_low):
        return pd.Series(["Ciencia y Análisis de Datos (Data Science, Analytics, SQL)", 'R'])
    if 'excel' in nombre_low:
        return pd.Series(["Ofimática y Productividad (Excel, Power BI, Office)", 'Excel'])
    if 'power bi' in nombre_low:
        return pd.Series(["Ciencia y Análisis de Datos (Data Science, Analytics, SQL)", 'Power BI'])
    if any(kw in nombre_low for kw in ['sql', 'database', 'base de datos', 'mysql']):
        return pd.Series(["Ciencia y Análisis de Datos (Data Science, Analytics, SQL)", 'SQL / Databases'])
    if re.search(r'\bgo\b', nombre_low):
        return pd.Series(["Desarrollo de Software y Web", 'Go (Golang)'])

    # 3. CAMPOS DE ESTUDIO (Rescate de lo que no especificó herramienta)
    herramienta = "General"
    if any(kw in nombre_low for kw in ['ciencia de datos', 'data science', 'bayesianos']):
        herramienta = 'Ciencia de Datos'
    elif any(kw in nombre_low for kw in [
        'analytics', 'tableau', 'lookml', 'big data', 'limpieza de datos',
        'serie temporal', 'visualización', 'seo', 'marketing', 'analítica',
        'pandas', 'características'
    ]):
        herramienta = 'Data Analytics / BI'

    # 4. DESARROLLO, INFRAESTRUCTURA Y DISEÑO
    elif any(kw in nombre_low for kw in ['html', 'css', 'front-end', 'web design', 'responsivo', 'angular', 'scratch', 'javascript', 'java', 'c++', 'c#', 'programación', 'coding', 'web', 'pila completa']):
        herramienta = 'Desarrollo de Software / Web'
    elif any(kw in nombre_low for kw in ['flutter', 'android', 'kotlin', 'jetpack', 'firebase']): herramienta = 'Desarrollo Móvil'
    elif any(kw in nombre_low for kw in ['cyber', 'security', 'seguridad', 'ciberseguridad']): herramienta = 'Ciberseguridad'
    elif any(kw in nombre_low for kw in ['cloud', 'nube', 'azure', 'aws', 'sre', 'api', 'linux', 'iot', 'redes']): herramienta = 'Infraestructura y Redes'

    # 5. OTROS ESPECIALIZADOS
    elif any(kw in nombre_low for kw in ['matlab', 'onramp', 'simulink', 'stateflow', 'wolfram', 'mathematica', 'cuántica', 'computación']):
        herramienta = 'Computación Científica'
    elif any(kw in nombre_low for kw in ['ux', 'usuario', 'diseño']): herramienta = 'Diseño UX/UI'
    elif any(kw in nombre_low for kw in ['blockchain', 'nft', 'crypto']): herramienta = 'Blockchain'
    elif any(kw in nombre_low for kw in ['word', 'powerpoint', 'office', 'computación fácil']): herramienta = 'Microsoft Office'
    elif any(kw in nombre_low for kw in ['inglés', 'english', 'basic a1']): herramienta = 'Inglés Técnico'
    elif any(kw in nombre_low for kw in ['liderazgo', 'leadership', 'conversations', 'liderar', 'comunicadores']): herramienta = 'Habilidades Blandas'

    # CATEGORÍA RELACIÓN (Mapeo Final de lo que quedó)
    if herramienta in ['Ciencia de Datos', 'Data Analytics / BI']:
        categoria = "Ciencia y Análisis de Datos (Data Science, Analytics, SQL)"
    elif herramienta in ['Ciberseguridad', 'Infraestructura y Redes']:
        categoria = "Redes, Ciberseguridad e Infraestructura"
    elif herramienta in ['Microsoft Office', 'Inglés Técnico', 'Habilidades Blandas', 'Blockchain']:
        categoria = "Ofimática y Productividad (Excel, Power BI, Office)"
    elif herramienta in ['Desarrollo de Software / Web', 'Desarrollo Móvil', 'Diseño UX/UI']:
        categoria = "Desarrollo de Software y Web"
    else:
        categoria = "STEM y Ciencias Básicas (Cálculo, Física, Programación básica)"

    return pd.Series([categoria, herramienta])

# --- PROCESAMIENTO DEL DATASET ROBOT ---
df_siev_robot = df_raw.iloc[17:198].copy()
df_siev_robot.columns = ['Curso_Nombre', 'URL_Acceso', 'Horas_Curso']

# Mapeo de Entidades
mapeo_entidades = {(0,6): "Cisco", (6,16): "CodeSignal", (16,22): "freeCodeCamp", (22,62): "Google", (62,71): "HackerRank", (71,81): "Harvard University", (81,85): "Hasso Plattner Institute", (85,89): "Hugging Face", (89,94): "IBM", (94,109): "Kaggle", (109,114): "LinkedIn Learning", (114,131): "MATLAB", (131,135): "Semrush Academy", (135,140): "University of Helsinki", (140,149): "upGrad", (149,158): "FutureLearn", (158,181): "Wolfram U"}
df_siev_robot['Entidad_Certifica'] = "Sin Clasificar"
for (i, f), ent in mapeo_entidades.items():
    df_siev_robot.iloc[i:f, df_siev_robot.columns.get_loc('Entidad_Certifica')] = ent

# LIMPIEZA Y TRADUCCIÓN
print("Iniciando procesamiento de limpieza y traducción avanzada...")
translator = GoogleTranslator(source='auto', target='es')

# 1. Filtro radical: Si no hay números en horas, no es tiempo (Limpia 'Baja demanda')
df_siev_robot['Horas_Curso'] = df_siev_robot['Horas_Curso'].astype(str).apply(lambda x: "No especifica" if not any(c.isdigit() for c in x) else x)

# 2. Traducción con Escudo de Términos (Python se mantiene)
df_siev_robot['Curso_Nombre'] = df_siev_robot['Curso_Nombre'].apply(lambda x: corregir_terminologia(translator.translate(str(x))) if pd.notnull(x) else x)
df_siev_robot['Horas_Curso'] = df_siev_robot['Horas_Curso'].apply(lambda x: translator.translate(str(x)) if x != "No especifica" else x)

# 3. Aplicar motor de clasificación final
df_siev_robot[['Categoria_Relacion', 'Herramienta_Principal']] = df_siev_robot['Curso_Nombre'].apply(clasificar_datos_pro)
df_siev_robot['Fuente'] = "Class Central"

print("Estado: Transformación completada satisfactoriamente.")
```
